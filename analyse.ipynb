# Analyse du dataset - Projet Shadow Client

## Objectif
Ce notebook présente l'analyse du dataset `cleaned_dataset.csv`. 
Il contient :
- Le chargement et l'exploration des données
- Les visualisations
- L'analyse descriptive et avancée
- Les insights principaux


```python
# ==============================
# 0️⃣ Import et préparation
# ==============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")

# ------------------------------
# Vérifier dossier et fichiers
# ------------------------------
print("Dossier courant :", os.getcwd())
print("Fichiers disponibles :", os.listdir())

# ==============================
# 1️⃣ Standardiser les colonnes
# ==============================
df = pd.read_csv("data/cleaned_dataset.csv")

df.rename(columns=lambda x: x.strip().lower().replace(" ", "_").replace("é","e").replace("è","e"), inplace=True)
print("\nColonnes après standardisation :")
print(df.columns.tolist())

# ==============================
# 2️⃣ Exploration du dataset
# ==============================
print("\nAperçu des premières lignes :")
display(df.head())

print("\nInfos sur le dataset :")
print(df.info())

print("\nStatistiques descriptives :")
display(df.describe())

# Colonnes numériques et catégorielles
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(include='object').columns
print("\nColonnes numériques :", num_cols)
print("Colonnes catégorielles :", cat_cols)

# Valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(df.isnull().sum())

# Valeurs uniques pour les colonnes catégorielles
for col in cat_cols:
    print(f"{col} : {df[col].nunique()} valeurs uniques")
    print(df[col].unique()[:10])
    print("-----")

# ==============================
# 3️⃣ Visualisations
# ==============================
# Histogrammes
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel("Nombre d'occurrences")
    plt.show()

# Heatmap des corrélations
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Corrélation entre colonnes numériques")
plt.show()

# Barplots pour colonnes catégorielles
for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index[:10], palette="pastel")
    plt.title(f"Répartition des 10 principales valeurs de {col}")
    plt.xticks(rotation=45)
    plt.ylabel("Nombre d'occurrences")
    plt.show()

# ==============================
# 4️⃣ Analyse avancée / insights
# ==============================
# Totaux et indicateurs
total_profit = df['profit'].sum() if 'profit' in df.columns else 'N/A'
total_ventes = df['quantite'].sum() if 'quantite' in df.columns else 'N/A'
total_clients = df['clientid'].nunique() if 'clientid' in df.columns else 'N/A'

print("📌 Indicateurs clés :")
print(f"Total des ventes : {total_ventes}")
print(f"Total des profits : {total_profit}")
print(f"Nombre de clients uniques : {total_clients}")

# Ventes et profits par catégorie
if 'categorie' in df.columns:
    ventes_par_categorie = df.groupby('categorie')['quantite'].sum() if 'quantite' in df.columns else None
    profit_par_categorie = df.groupby('categorie')['profit'].sum()
    
    print("\nProfits par catégorie :")
    display(profit_par_categorie.sort_values(ascending=False))
    
    # Visualisation
    plt.figure(figsize=(8,4))
    profit_par_categorie.sort_values(ascending=False).plot(kind='bar', color='salmon')
    plt.title("Profit par catégorie")
    plt.ylabel("Profit total")
    plt.show()

# Identifier les anomalies
if 'profit' in df.columns:
    profits_negatifs = df[df['profit'] < 0]
    print(f"\nNombre de transactions avec profit négatif : {len(profits_negatifs)}")
    display(profits_negatifs.head())

# Corrélation prix-unitaire et quantite (si colonnes présentes)
if 'prix-unitaire' in df.columns and 'quantite' in df.columns:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x='prix-unitaire', y='quantite')
    plt.title("Corrélation entre prix-unitaire et quantité vendue")
    plt.xlabel("Prix-unitaire")
    plt.ylabel("Quantité")
    plt.show()

# ==============================
# 5️⃣ Résumé Markdown dynamique
# ==============================
markdown_summary = f"""
# Résumé et Insights – Projet Shadow Client

## 1️⃣ Indicateurs clés
- Total des profits : {total_profit}
- Total des ventes : {total_ventes} (non disponible si N/A)
- Nombre de clients uniques : {total_clients} (non disponible si N/A)

## 2️⃣ Analyse par catégorie
- Profits par catégorie : voir visualisation ci-dessus
- Ventes par catégorie : voir visualisation ci-dessus (si disponible)

## 3️⃣ Anomalies et points d’attention
- Transactions avec profit négatif : {len(profits_negatifs) if 'profits_negatifs' in locals() else 'N/A'}

## 4️⃣ Tendances et corrélations
- Corrélation entre prix-unitaire et quantité vendue : voir graphique ci-dessus

## 5️⃣ Conclusion
- Identification des produits et catégories clés pour la rentabilité
- Optimisation possible des prix et actions sur les produits performants
- Prochaines étapes : segmentation, prévision du profit, ou analyse plus détaillée
"""

print(markdown_summary)
```
